```python
from randomuser import RandomUser
import pandas as pd
```

```python
randomUser = RandomUser()
```

```python
userList = randomUser.generate_users(10)
userList
```

```python
for thisUser in userList:
    print(thisUser.get_full_name() + ", " + thisUser.get_email())
```

```python
for thisUser in userList:
    print(thisUser.get_picture())
```

```python
def getUsers():
    userData = []
    for thisUser in RandomUser.generate_users(10):
        userData.append({"Name": thisUser.get_full_name(),
                        "Gender": thisUser.get_gender(),
                        "City": thisUser.get_city(),
                        "State": thisUser.get_state(),
                        "Email": thisUser.get_email(),
                        "DOB": thisUser.get_dob(),
                        "Picture": thisUser.get_picture()})
    return userData
df1 = pd.DataFrame(getUsers())
df1
```

```python
import requests
import json
```

```python
fvData = requests.get("https://fruityvice.com/api/fruit/all")
```

```python
fvRes = json.loads(fvData.text)
```

```python
df2 = pd.json_normalize(fvRes)
df2
```

```python
cherry = df2.loc[df2["name"] == "Cherry"]
(cherry.iloc[0]["family"], cherry.iloc[0]["genus"])
```

```python
banana = df2.loc[df2["name"] == "Banana"]
banana.iloc[0]["nutritions.calories"]
```
